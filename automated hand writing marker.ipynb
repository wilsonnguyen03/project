```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
import easyocr
```

```python
# Load the image
image_path = "C:/Users/Admin/Desktop/University/UQ/Coding/Handwriting project/1.jpg"
image = cv2.imread(image_path)

# Convert BGR to RGB (for proper display)
#image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
# Convert image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to remove noise
_, thresh = cv2.threshold(gray, 140, 255, cv2.THRESH_BINARY_INV)

# Show processed image
plt.imshow(thresh, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Set Tesseract path if needed (Windows only)

pytesseract.pytesseract.tesseract_cmd = r"C:/Program Files/Tesseract-OCR/tesseract.exe"

# Extract text
extracted_text = pytesseract.image_to_string(thresh, config="--psm 6")

print("Extracted Text:\n", extracted_text)
```

```python
# Initialize the OCR reader
reader = easyocr.Reader(["en"])  # "en" for English

# Perform OCR
results = reader.readtext(image)

# Print detected text
for (bbox, text, prob) in results:
    print(f"Detected: {text} (Confidence: {prob:.2f})")
```
